```python
import pandas as pd
from irrCAC.raw import CAC
```

GWET interpretation:
Gwet's first-order agreement coefficient (Gwet's AC1) can be interpreted using Landis & Koch's rule of thumb. Therefore, Gwet's AC1 values of < 0.00, 0.00 to 0.20, 0.21 to 0.40, 0.41 to 0.60, 0.61 to 0.80, and 0.81 to 100 indicate poor, slight, fair, moderate, substantial, and almost perfect agreement levels respectively

```python
data_path = "/mnt/data1/klug/datasets/meditron/Meditron_ICU.xlsx"
```

```python
df = pd.read_excel(data_path)
```

```python
# remap Votes : 1 -> -1 / 2 -> +1 / 12 -> 0
df["Vote"] = df["Vote"].map({1: -1, 2: +1, 12: 0})
```

```python
# df groupby question then see Name as columns and Votes as values
df_pivot = df.pivot_table(index='Question', columns='Name', values='Vote')
df_pivot
```

```python

cac = CAC(df_pivot)
```

```python
cac.fleiss()
```

```python
cac.krippendorff()
```
